```python
import unicodedata
def limpiar_nombre(nombre):
    nombre = nombre.lower()
    nombre = unicodedata.normalize('NFKD', nombre).encode('ASCII', 'ignore').decode('utf-8')
    nombre = nombre.strip().replace(' ', '_')
    return nombre
```

```python
df.columns = [limpiar_nombre(col) for col in df.columns]
```
